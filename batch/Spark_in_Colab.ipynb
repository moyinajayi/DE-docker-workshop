```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
        .appName('testColab') \
        .getOrCreate()
```

```python
df = spark.read.parquet('yellow_tripdata_2025-11')
df.count()
```

```python
from pyspark.sql.functions import to_date, col

# Count trips that started on November 15th
df.filter(to_date(col('tpep_pickup_datetime')) == '2025-11-15').count()
```

```python
from pyspark.sql.functions import col, unix_timestamp

# Show the longest trip details
df.withColumn(
    'trip_hours', 
    (unix_timestamp('tpep_dropoff_datetime') - unix_timestamp('tpep_pickup_datetime')) / 3600
).orderBy(col('trip_hours').desc()).select(
    'tpep_pickup_datetime', 
    'tpep_dropoff_datetime', 
    'trip_hours'
).show(5)
```

```python
!curl -O https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv
```

```python
zones_df = spark.read.csv('taxi_zone_lookup.csv', header=True, inferSchema=True)
zones_df.createOrReplaceTempView('zones')
zones_df.show()
```

```python
from pyspark.sql.functions import col, count

df.join(zones_df, df.PULocationID == zones_df.LocationID) \
  .groupBy('Zone') \
  .agg(count('*').alias('pickup_count')) \
  .orderBy('pickup_count') \
  .show(1, truncate=False)
```

```python
# First, create temp view for trips if not already done
df.createOrReplaceTempView('trips')

# Find least frequent pickup location zone
spark.sql("""
    SELECT z.Zone, COUNT(*) as pickup_count
    FROM trips t
    JOIN zones z ON t.PULocationID = z.LocationID
    GROUP BY z.Zone
    ORDER BY pickup_count ASC
    LIMIT 1
""").show(truncate=False)
```

```python
# Check data types
print("PULocationID type:", df.schema['PULocationID'].dataType)
print("LocationID type:", zones_df.schema['LocationID'].dataType)
```

```python
# Re-create temp view from current df
df.createOrReplaceTempView('trips')

# Compare counts
print("df count:", df.count())
print("trips view count:", spark.sql("SELECT COUNT(*) FROM trips").collect()[0][0])
```

```python
print("\nDataFrame API result:")
df.join(zones_df, df.PULocationID == zones_df.LocationID) \
  .groupBy('Zone') \
  .agg(count('*').alias('pickup_count')) \
  .orderBy('pickup_count') \
  .show(1, truncate=False)
```

```python
print("\nSQL result:")
spark.sql("""
    SELECT z.Zone, COUNT(*) as pickup_count
    FROM trips t
    JOIN zones z ON t.PULocationID = z.LocationID
    GROUP BY z.Zone
    ORDER BY pickup_count ASC
    LIMIT 1
""").show(truncate=False)
```

```python
spark.sql("""
    SELECT z.Zone, COUNT(*) as pickup_count
    FROM trips t
    JOIN zones z ON t.PULocationID = z.LocationID
    GROUP BY z.Zone
    HAVING COUNT(*) = 1
    ORDER BY z.Zone
""").show(truncate=False)
```

```python
df.join(zones_df, df.PULocationID == zones_df.LocationID) \
  .groupBy('Zone') \
  .agg(count('*').alias('pickup_count')) \
  .orderBy('pickup_count', 'Zone') \
  .show(10, truncate=False)
```
